###ventas_precios_corrientes	



```python
#Realizamos histograma de variables cuantitativas
fig = plt.figure(figsize =(5, 3))
plt.hist(x=data2["ventas_precios_corrientes"], color='blue', rwidth=0.50)
plt.title('Histograma de Precio de ventas')
plt.xlabel('Ventas')
plt.ylabel('Frecuencia')

plt.show() #dibujamos el histograma
```

```python
#Realizamos diagrama de caja o bigote
fig = plt.figure(figsize =(10, 6))
plt.boxplot(data2["ventas_precios_corrientes"]) 
plt.title("Outliers de  Ventas totales ")
plt.show() #dibujamos el diagrama
```

**MÃ©todo de desviaciÃ³n estÃ¡ndar**


```python
#MÃ©todo aplicando desviaciÃ³n estandar. Encuentro los valores extremos
y=data2["ventas_precios_corrientes"]
Limite_Superior= y.mean() + 3*y.std()
Limite_Inferior= y.mean() - 3*y.std()
print("Limite superior permitido", Limite_Superior)
print("Limite inferior permitido", Limite_Inferior)
```

```python
#Encontramos Ouliers
outliers= data2[(y>Limite_Superior)|(y<Limite_Inferior)]
outliers
```

```python
#Obtenemos datos limpios
data_clean= data2[(y<=Limite_Superior)&(y>=Limite_Inferior)]
data_clean
```

```python
#Realizamos diagrama de caja o bigote
fig = plt.figure(figsize =(5, 3))
plt.boxplot(data_clean["ventas_precios_corrientes"]) 
plt.title("Outliers de Precio Venta")
plt.show() #dibujamos el diagrama
```

**MÃ‰TODO DE IQR (RANGO INTER QUARTILICO)**

```python
#MÃ©todo aplicando Cuartiles. Encuentro cuartiles 0.25 y 0.75
y=data2["ventas_precios_corrientes"]

percentile25=y.quantile(0.25) #Q1
percentile75=y.quantile(0.75) #Q3
iqr= percentile75 - percentile25

Limite_Superior_iqr= percentile75 + 1.5*iqr
Limite_Inferior_iqr= percentile25 - 1.5*iqr
print("Limite superior permitido", Limite_Superior_iqr)
print("Limite inferior permitido", Limite_Inferior_iqr)
```

```python
#Encontramos Ouliers
outliers_iqr= data2[(y>Limite_Superior_iqr)|(y<Limite_Inferior_iqr)]
outliers_iqr
```

```python
#Obtenemos datos limpios
data_clean_iqr= data2[(y<=Limite_Superior_iqr)&(y>=Limite_Inferior_iqr)]
data_clean_iqr
```

```python
#Realizamos diagrama de caja o bigote
fig = plt.figure(figsize =(5, 3))
plt.boxplot(data_clean_iqr["ventas_precios_corrientes"]) 
plt.title("Outliers de  Ventas totales")
plt.show() #dibujamos el diagrama
```

```python
#Convertir archivo filtrado a CSV
data_clean_iqr.to_csv("outliers_ventas_precios_constantes.csv")
```
